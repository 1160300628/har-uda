```python
import seaborn as sns

def plot_conf(conf_mat):
    df = pd.DataFrame(conf_mat)
    df.index = ['WALKING', 'WALKING_UPSTAIRS', 'WALKING_DOWNSTAIRS', 'SITTING', 'STANDING', 'LAYING']
    df.columns = ['WALKING', 'WALKING_UPSTAIRS', 'WALKING_DOWNSTAIRS', 'SITTING', 'STANDING', 'LAYING']
    sns.heatmap(df, annot=True, cmap='RdBu')
```

```python
conf_4 = conf_mat[4][4].astype(int)
plot_conf(conf_4)
```

```python
conf_6 = conf_mat[6][6].astype(int)
plot_conf(conf_6)
```

```python
conf_8 = conf_mat[8][8].astype(int)
plot_conf(conf_8)
```
